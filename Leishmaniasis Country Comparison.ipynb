```python
grp=lesh_data.groupby("Country")
x=grp["Total number of cases since 2005"].agg(np.mean)
```

```python
#since it is a little bit messy
plt.figure(figsize=(16,5))
#to make it bigger (length,height)
plt.plot(x)
plt.xticks(rotation=90)#to fix the x 
#Add some titles according to variable x 
plt.title("Number of cases in the last 15 years of cutaneous leishmaniasis reported data by country")
plt.xlabel("Countries-->")
plt.ylabel("Number of cases reported-->")
plt.show()
```
